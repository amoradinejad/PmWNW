```python
pknw_interp_dst =  InterpolatedUnivariateSpline(k_input,pnw_dst,k=3)
pknw_interp_g =  InterpolatedUnivariateSpline(k_input,pnw_Gfilter,k=3)
pknw_interp   =  InterpolatedUnivariateSpline(k_input,pnw_Bspline,k=3)
pknw_dst      = pknw_interp_dst(k_input)
pkw_dst       = plin - pknw_dst
pknw_g      = pknw_interp_g(k_input)
pkw_g       = plin - pknw_g
pknw        = pknw_interp(k_input)
pkw         = (plin - pknw)

def F2_s(k,q,mu):
    f = q**2.*(3.*k + 7.*q*mu - 10.*k*mu**2.)/(14.*k*(k**2. + q**2. - 2.*k*q*mu));
    return f

def F3_s(k,q,mu):
    kmq2   = np.abs(q**2.+k**2.-2.*q*k*mu);
    kdotq = k*q*mu;
    f = 1./kmq2*(5*k**2./63.-(11.*kdotq)/54.-k**2.*kdotq**2./(6.*q**4.)+19.*kdotq**3./(63.*q**4.)-23.*k**2.*kdotq/(378.*q**2.)-23.*kdotq**2./(378.*q**2.)+kdotq**3./(9.*k**2*q**2.));
    return f
            
def p22_integrand(logq, mu, k):
    q = np.exp(logq)
    kmq = np.sqrt(np.abs(q**2.+k**2.-2.*q*k*mu));
    f2s = F2_s(q, k, mu)
    plin_q = pk_interp(q)
    plin_kmq = pk_interp(kmq)
    f = 2.*2.*np.pi*q**3*plin_q*plin_kmq*f2s**2.;
    return f
    
def p13_integrand(logq, mu, k):
    q = np.exp(logq)
    plin_k = pk_interp(k)
    plin_q = pk_interp(q)
    f = 6.*2.*np.pi*q**3*plin_k*plin_q*F3_s(k, q, mu);
    return f

def p22(k):
    f = dblquad(p22_integrand,-1.,1., lambda mu:np.log(1.e-4),lambda mu:np.log(100.), args=(k,),epsrel = 1.e-3)[0]
    return f

def p13(k):
    f = dblquad(p13_integrand,-1.,1. ,lambda mu:np.log(1.e-4),lambda mu: np.log(100.),args=(k,),epsrel = 1.e-3)[0]
    return f

vp22 = np.vectorize(p22)
vp13 = np.vectorize(p13)

matter_sim_file = 'avgps_mm_z0.0_byagox_gaussian_1536_2Gpc_085.dat'
k, avek, aveP, s_aveP = np.loadtxt(matter_sim_file, unpack= True)

# plt.plot(k_input,pk_interp(k_input), label = 'linear')
# plt.scatter(avek,aveP, label ='measurment')
# plt.errorbar(avek,aveP,yerr=s_aveP, fmt = 'none')
# plt.xscale('log')
# plt.yscale('log')
# plt.xlim(5.e-3, 1.)
# plt.ylim(1.,200.)
# plt.legend()
# plt.savefig("ps_matter.png")
# plt.show()

# plt.scatter(avek,pk_interp(avek)-aveP, label = 'linear')
# plt.xscale('log')
# plt.yscale('linear')
# plt.xlim(3.e-3,0.5)
# plt.show()
# plt.loglog(k_input, vp22(k_input),label = 'P22')
# plt.loglog(k_input, vp13(k_input),label = 'P13')
# plt.loglog(k_input, pk_interp(k_input),label = 'lin')
# plt.show()

k1, IR_LO1, IR_NLO1 = np.loadtxt("IR_ks_p05.txt",unpack =True)
k2, IR_LO2, IR_NLO2 = np.loadtxt("IR_ks_p1.txt",unpack =True)
k3, IR_LO3, IR_NLO3 = np.loadtxt("IR_ks_p15.txt",unpack =True)
k4, IR_LO4, IR_NLO4 = np.loadtxt("IR_ks_p2_Bspline.txt",unpack =True)
k4_g, IR_LO4_g, IR_NLO4_g = np.loadtxt("IR_ks_p2_gfilter.txt",unpack =True)



# plt.semilogx(k1, pk_interp(k1)/pknw_interp(k1), ls='--',label='lin')
# plt.semilogx(k1,IR_LO1/pknw_interp(k1),label ='$ k_s= 0.05$')
# plt.semilogx(k2,IR_LO2/pknw_interp(k1),label ='$ k_s= 0.1$')
# plt.semilogx(k3,IR_LO3/pknw_interp(k1),label ='$ k_s= 0.15$')
# plt.semilogx(k4,IR_LO4/pknw_interp(k4),label ='$ k_s= 0.2$')
# plt.text(0.01, 1.03, 'IR-res, LO')
# plt.xlabel('$k \; [h/\mathrm{Mpc}]$')
# plt.ylabel('$P(k)/P_\mathrm{nw}(k)$')
# plt.legend()
# plt.xlim(5.e-3,1.)
# plt.ylim(0.93,1.075)
# plt.legend()
# plt.savefig("IR_LO_ks.png")
# plt.show()


# plt.semilogx(k1, pk_interp(k1)/pknw_interp(k1), ls='--',label='lin')
# plt.semilogx(k1,IR_NLO1/pknw_interp(k1),label ='$ k_s= 0.05$')
# plt.semilogx(k2,IR_NLO2/pknw_interp(k1),label ='$ k_s= 0.1$')
# plt.semilogx(k3,IR_NLO3/pknw_interp(k1),label ='$ k_s= 0.15$')
# plt.semilogx(k4,IR_NLO4/pknw_interp(k4),label ='$ k_s= 0.2$')
# plt.text(0.01, 1.03, 'IR-res, NLO')
# plt.xlabel('$k \; [h/\mathrm{Mpc}]$')
# plt.ylabel('$P(k)/P_\mathrm{nw}(k)$')
# plt.legend()
# plt.xlim(5.e-3,1.)
# plt.ylim(0.93,1.075)
# plt.legend()
# plt.savefig("IR_NLO_ks.png")
# plt.show()


IR_NLO_interp = InterpolatedUnivariateSpline(k4,IR_NLO4,k=3)
IR_LO_interp = InterpolatedUnivariateSpline(k4,IR_LO4,k=3)

pk_NLO_lin = (IR_NLO_interp(avek)-2.*1.9*avek**2.*pk_interp(avek))
pk_NLO_LO  = (IR_NLO_interp(avek)-2.*1.9*avek**2.*IR_LO_interp(avek))
pk_LO_lin  = (IR_LO_interp(avek) -2.*1.9*avek**2.*pk_interp(avek))
pk_LO_LO  = (1. -2.*1.9*avek**2.)*IR_LO_interp(avek)


plt.semilogx(avek, pk_interp(avek)/pknw_interp(avek),ls='--',label='lin')
plt.semilogx(avek, pk_LO_LO/pknw_interp(avek),label='LO',c = "purple")
plt.semilogx(avek, pk_NLO_LO/pknw_interp(avek),label ='NLO',c = 'r')
plt.scatter(avek,aveP/pknw_interp(avek),s=20)
plt.errorbar(avek,aveP/pknw_interp(avek),yerr=s_aveP/pknw_interp(avek), fmt = 'none')
plt.xlim(3.e-3,0.5)
plt.ylim(0.82,1.12)
plt.xlabel('$k \; [h/\mathrm{Mpc}]$')
plt.ylabel('$P(k)/P_\mathrm{nw}(k)$')
plt.legend(loc=4)
plt.savefig('IR_LO_NLO_sims_ct.png')
plt.show()

# plt.semilogx(avek,(pk_NLO_LO/pk_NLO_lin -1.)*100)
# plt.xlabel('$k \; [h/\mathrm{Mpc}]$')
# plt.ylabel('$(P_\mathrm{NLO}[P_\mathrm{ct, LO}])/P_\mathrm{NLO}[P_\mathrm{ct, lin}]  - 1 )\%$')
# plt.xlim(1.e-2,1.)
# plt.ylim(-.4,0.4)
# #plt.savefig("ct_diff.png")
# plt.show()

# plt.semilogx(avek, pk_interp(avek)/pknw_interp(avek),ls='--',label='lin')
# plt.semilogx(avek, pk_NLO_lin/pknw_interp(avek),c = "purple", label = "$-2 c_s^2 k^2 P_\mathrm{lin}$")
# plt.semilogx(avek, pk_NLO_LO/pknw_interp(avek),c = 'r', label = "$-2 c_s^2 k^2P_\mathrm{LO}$")
# plt.scatter(avek,aveP/pknw_interp(avek))
# plt.errorbar(avek,aveP/pknw_interp(avek),yerr=s_aveP/pknw_interp(avek), fmt = 'none')
# plt.xlim(2.e-3,0.5)
# plt.ylim(0.9,1.1)
# plt.xlabel('$k \; [h/\mathrm{Mpc}]$')
# plt.ylabel('$P(k)/P_\mathrm{nw}(k)$')
# plt.legend()
# #plt.savefig("ct_comp.png")
# plt.show()


# plt.semilogx(k1, pk_interp(k1)/pknw_interp(k1), ls='--',label='lin')
# plt.semilogx(k4,IR_LO4/pknw_interp(k4),label ='LO-Bspline', c = "purple")
# plt.semilogx(k4,IR_NLO4/pknw_interp(k4),label ='NLO-Bspline',c = 'r')
# plt.xlim(3.e-3,1.)
# plt.ylim(0.9,1.1)
# plt.xlabel('$k \; [h/\mathrm{Mpc}]$')
# plt.ylabel('$P(k)/P_\mathrm{nw}(k)$')
# plt.legend()
# #plt.savefig("IR_LO_NLO.png")
# plt.show()

# plt.semilogx(k1, pk_interp(k1)/pknw_interp(k1), ls='--',label='lin')
# plt.semilogx(k4,IR_LO4/pknw_interp(k4),label ='LO-Bspline', c = "purple")
# plt.semilogx(k4,IR_NLO4/pknw_interp(k4),label ='NLO-Bspline',c = 'r')
# plt.semilogx(k4,IR_LO4_g/pknw_interp_g(k4),label ='LO-Gfilter', ls='--', c = "purple")
# plt.semilogx(k4,IR_NLO4_g/pknw_interp_g(k4),label ='NLO-Gfilter',ls ='--', c = 'r')
# plt.xlim(3.e-3,1.)
# plt.ylim(0.9,1.1)
# plt.xlabel('$k \; [h/\mathrm{Mpc}]$')
# plt.ylabel('$P(k)/P_\mathrm{nw}(k)$')
# plt.legend()
# plt.savefig("IR_LO_NLO_Bspline_gfilter.png")
# plt.show()

plt.scatter(avek, 100.*(aveP/pk_interp(avek)-1.),s=20,label='lin')
plt.errorbar(avek,100.*(aveP/pk_interp(avek)-1.),s=20,yerr=100.*s_aveP/pk_interp(avek), fmt = 'none', color='#ff7f0e')
plt.scatter(avek, 100.*(aveP/pk_LO_LO-1.),s=20,label='LO',c = "purple")
plt.scatter(avek, 100.*(aveP/pk_NLO_LO-1.),s=20,label ='NLO',c = 'r')
plt.xlim(3.e-3,0.5)
plt.ylim(-19,18)
plt.xscale('log')
plt.xlabel('$k \; [h/\mathrm{Mpc}]$')
plt.ylabel('$(P_\mathrm{sims}(k)/P_\mathrm{th}(k) - 1) \% $')
plt.legend(loc=4)
plt.savefig('psim_pth_ratio_ct_dp.png')
plt.show()

def sigma2_IR_integrand_dst(x):
    k_osc = 1./110.
    qOk = x/k_osc
    integ = 4.*np.pi/3.*pknw_interp_dst(x)*(1.-spherical_jn(0, qOk) + 2.*spherical_jn(2, qOk))
    return integ

def sigma2_IR_dst(ks):
    sig2 = quad(sigma2_IR_integrand,1.e-6, ks, epsrel = 1.e-4, limit=10000)[0]
    return sig2

def P_LO_dst(k,ks):
    pkww  = (pk_interp(k)-pknw_interp_dst(k))
    pknww = pknw_interp_dst(k)
    return pknww + np.exp(-k**2.*sigma2_IR_dst(ks))* pkww

def sigma2_IR_integrand_g(x):
    k_osc = 1./110.
    qOk = x/k_osc
    integ = 4.*np.pi/3.*pknw_interp_g(x)*(1.-spherical_jn(0, qOk) + 2.*spherical_jn(2, qOk))
    return integ

def sigma2_IR_g(ks):
    sig2 = quad(sigma2_IR_integrand,1.e-6, ks, epsrel = 1.e-4, limit=10000)[0]
    return sig2

def P_LO_g(k,ks):
    pkww  = (pk_interp(k)-pknw_interp_g(k))
    pknww = pknw_interp_g(k)
    return pknww + np.exp(-k**2.*sigma2_IR_g(ks))* pkww


def sigma2_IR_integrand(x):
    k_osc = 1./110.
    qOk = x/k_osc
    integ = 4.*np.pi/3.*pknw_interp(x)*(1.-spherical_jn(0, qOk) + 2.*spherical_jn(2, qOk))
    return integ

def sigma2_IR(ks):
    sig2 = quad(sigma2_IR_integrand,1.e-6, ks, epsrel = 1.e-4, limit=10000)[0]
    return sig2

def P_LO(k,ks):
    pkww  = (pk_interp(k)-pknw_interp(k))
    pknww = pknw_interp(k)
    return pknww + np.exp(-k**2.*sigma2_IR(ks))* pkww

plt.semilogx(k_input, plin/pknw, ls='--',label='lin')
plt.semilogx(k_input, P_LO(k_input,.2)/pknw,label='Bspline')
plt.semilogx(k_input, P_LO_g(k_input,.2)/pknw_g, label='Gfilter')
#plt.semilogx(k_input, P_LO_dst(k_input,.2)/pknw_dst, label='spectral')
plt.text(0.01, 1.03, 'IR-res, LO')
plt.xlim(2.e-3,1.)
plt.ylim(0.93,1.075)
plt.xlabel('$k \; [h/\mathrm{Mpc}]$')
plt.ylabel('$P(k)/P_\mathrm{nw}(k)$')
plt.legend()
#plt.savefig("IR_LO_Msplit.png")
plt.show()

# plt.semilogx(k_input, (P_LO_g(k_input,.2)-P_LO(k_input,.2)), label='Gfilter')
# #plt.semilogx(k_input, P_LO_dst(k_input,.2)/pknw_dst, label='spectral')
# #plt.text(0.01, 1.03, 'IR-res, LO')
# # plt.xlim(2.e-3,1.)
# # plt.ylim(0.93,1.075)
# plt.xlabel('$k \; [h/\mathrm{Mpc}]$')
# plt.ylabel('$P(k)/P_\mathrm{nw}(k)$')
# plt.legend()
# #plt.savefig("IR_LO_Msplit.png")
# plt.show()


# vlogk = np.linspace(np.log(1.e-3),np.log(1.),300)
# vk = np.exp(vlogk)
# vsigmaks = np.vectorize(sigma2_IR)
# plt.loglog(vk,vsigmaks(vk))
# plt.xlabel('$k_s \; [h/\mathrm{Mpc}]$')
# plt.ylabel('$\Sigma^2(k_s)\; [(\mathrm{Mpc}/h)^3]$')
# plt.xlim(1.e-3,1.)
# plt.ylim(6.e-3,100.)
# plt.savefig("IR_sig2.png")
# plt.show()


# plt.semilogx(k_input, plin/pknw, ls='--',label='lin')
# plt.semilogx(k_input, P_LO(k_input,.05)/pknw,label='$ k_s= 0.05$')
# plt.semilogx(k_input, P_LO(k_input,.1)/pknw, label='$ k_s= 0.1$')
# plt.semilogx(k_input, P_LO(k_input,.15)/pknw, label='$k_s= 0.15$')
# plt.semilogx(k_input, P_LO(k_input,.2)/pknw,  label='$k_s= 0.2$')
# plt.text(0.01, 1.03, 'IR-res, LO')
# plt.xlim(5.e-3,1.)
# plt.ylim(0.93,1.075)
# plt.xlabel('$k \; [h/\mathrm{Mpc}]$')
# plt.ylabel('$P(k)/P_\mathrm{nw}(k)$')
# plt.legend()
# plt.savefig("IR_LO_ks.png")
# plt.show()
```
