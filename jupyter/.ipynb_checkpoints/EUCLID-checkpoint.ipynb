```python
# plt.loglog(k_input, pknw_interp_g(k_input))
# plt.show()


pknw_interp_dst     =  InterpolatedUnivariateSpline(k_input,pnw_dst,k=3)
pknw_interp_g       =  InterpolatedUnivariateSpline(k_input,pnw_Gfilter,k=3)
pknw_interp_Bspline =  InterpolatedUnivariateSpline(k_input,pnw_Bspline,k=3)

pknw_dst        = pknw_interp_dst(k_input)
pkw_dst         = plin - pknw_dst

pknw_g          = pknw_interp_g(k_input)
pkw_g           = plin - pknw_g

pknw_Bspline    = pknw_interp_Bspline(k_input)
pkw_Bspline     = plin - pknw_Bspline

def sigma2_IR_integrand_dst(x):
    k_osc = 1./110.
    qOk = x/k_osc
    integ = 1./(6.*np.pi**2.)*pknw_interp_dst(x)*(1.-spherical_jn(0, qOk) + 2.*spherical_jn(2, qOk))
    return integ

def sigma2_IR_dst(ks):
    sig2 = quad(sigma2_IR_integrand_dst,1.e-3, ks, epsrel = 1.e-4, limit=10000)[0]
    return sig2

def P_LO_dst(k,ks):
    pkww  = pk_interp(k)-pknw_interp_dst(k)
    pknww = pknw_interp_dst(k)
    return pknww + np.exp(-k**2.*sigma2_IR_dst(ks))* pkww

def sigma2_IR_integrand_g(x):
#     k_osc = np.pi/110.
    k_osc = 1./110.
    qOk = x/k_osc
    integ = 1./(6.*np.pi**2.)*pknw_interp_g(x)*(1.-spherical_jn(0, qOk) + 2.*spherical_jn(2, qOk))
    return integ

def sigma2_IR_g(ks):
    sig2 = quad(sigma2_IR_integrand_g,1.e-6, ks, epsrel = 1.e-4, limit=100000)[0]
    return sig2

def P_LO_g(k,ks):
    pkww  = pk_interp(k)-pknw_interp_g(k)
    pknww = pknw_interp_g(k)
    PLO = pknww + np.exp(-k**2.*sigma2_IR_g(ks))* pkww
#     print(pknww, pkww, PLO)
    return PLO

def sigma2_IR_integrand_Bspline(x):
    k_osc = 1./110.
    qOk = x/k_osc
    integ = 1./(6.*np.pi**2.)*pknw_interp_Bspline(x)*(1.-spherical_jn(0, qOk) + 2.*spherical_jn(2, qOk))
    return integ

def sigma2_IR_Bspline(ks):
    sig2 = quad(sigma2_IR_integrand_Bspline,1.e-4, ks, epsrel = 1.e-4, limit=10000)[0]
    return sig2

def P_LO_Bspline(k,ks):
    pkww  = (pk_interp(k)-pknw_interp_Bspline(k))
    pknww = pknw_interp_Bspline(k)
    return pknww + np.exp(-k**2.*sigma2_IR_Bspline(ks))* pkww

# plt.semilogx(k_input, plin/pknw_interp_Bspline(k_input), ls='--',label='lin')
# plt.semilogx(k_input, P_LO_Bspline(k_input,.2)/pknw_interp_Bspline(k_input),label='Bspline')
# plt.semilogx(k_input, P_LO_g(k_input,.2)/pknw_interp_g(k_input), label='Gfilter')
# plt.semilogx(k_input, P_LO_dst(k_input,.2)/pknw_interp_dst(k_input), label='spectral')
# plt.text(0.005, 0.96, 'IR-res, LO')
# plt.xlim(1.e-3,10.)
# # plt.ylim(0.93,1.075)
# plt.xlabel('$k \; [h/\mathrm{Mpc}]$')
# plt.ylabel('$P_\mathrm{LO}(k)/P_\mathrm{nw}(k)$')
# plt.legend()
# plt.savefig("LO_each.pdf")
# plt.show()

# plt.semilogx(k_input, plin/pknw_interp_Bspline(k_input), ls='--',label='lin')
# plt.semilogx(k_input, P_LO_Bspline(k_input,.2)/pknw_interp_Bspline(k_input),label='Bspline')
# plt.semilogx(k_input, P_LO_g(k_input,.2)/pknw_interp_Bspline(k_input), label='Gfilter')
# plt.semilogx(k_input, P_LO_dst(k_input,.2)/pknw_interp_Bspline(k_input), label='spectral')
# plt.xlim(1.e-3,10.)
# plt.ylim([0.93,1.08])
# plt.xlabel('$k \; [h/\mathrm{Mpc}]$')
# plt.ylabel('$P_\mathrm{LO}(k)/P_\mathrm{nw}(k)$')
# plt.legend()
# plt.savefig("LO_Bspline.pdf")
# plt.show()

# plt.semilogx(k_input, 100.*(P_LO_Bspline(k_input,.2)/P_LO_g(k_input,.2)-1.), label="Bspline, Gfilter")
# plt.semilogx(k_input, 100.*(P_LO_Bspline(k_input,.2)/P_LO_dst(k_input,.2)-1.), label="Bspline, spectral")
# plt.semilogx(k_input, 100.*(P_LO_g(k_input,.2)/P_LO_dst(k_input,.2)-1.), label="Gfilter, spectral")
# plt.xlabel('$k \; [h/\mathrm{Mpc}]$')
# plt.ylabel('frac diff [\%]')
# #plt.text(0.005, -0.3, 'IR-res, LO')
# plt.xlim(1.e-3,10.)
# plt.ylim([-.75,0.75])
# plt.legend()
# plt.savefig("frac_diff_LO.pdf")
# plt.show()

# vsigmaks = np.vectorize(sigma2_IR_g)
# plt.loglog(k_input,vsigmaks(k_input))
# plt.xlabel('$k_s \; [h/\mathrm{Mpc}]$')
# plt.ylabel('$\Sigma^2(k_s)\; [(\mathrm{Mpc}/h)^3]$')
# plt.xlim(k_input[0],k_input[-1])
# plt.ylim(1.e-4,100.)
# plt.savefig("IR_sig2.png")
# plt.show()


# km, IR_LO_bspline, IR_LO_gfilter, IR_LO_dst, IR_NLO_bspline, IR_NLO_gfilter, IR_NLO_dst = np.loadtxt("IR_LONLO_comp.txt",unpack =True)
# plt.semilogx(km, PK_L.P(0,km)/pknw_interp_Bspline(km), ls='--')
# plt.semilogx(km, IR_LO_bspline/pknw_interp_Bspline(km),label='Bspline-LO')
# plt.semilogx(km, IR_LO_gfilter/pknw_interp_Bspline(km),label='Gfilter-LO')
# plt.semilogx(km, IR_LO_dst/pknw_interp_Bspline(km),label='DST-LO')
# plt.semilogx(km, IR_NLO_bspline/pknw_interp_Bspline(km),label='Bspline-NLO')
# plt.semilogx(km, IR_NLO_gfilter/pknw_interp_Bspline(km),label='Gfilter-NLO')
# plt.semilogx(km, IR_NLO_dst/pknw_interp_Bspline(km),label='DST-NLO', color='darkcyan')
# plt.xlim(1.e-3,1.)
# plt.ylim(0.93,1.08)
# plt.minorticks_on()
# plt.xlabel('$k \; [h/\mathrm{Mpc}]$')
# plt.ylabel('$P(k)/P_\mathrm{nw-Bspline}(k)$')
# plt.legend(loc = 2, ncol=2)
# plt.savefig("LO_NLO_all.pdf")
# plt.show()

# plt.semilogx(km,100.*(IR_LO_bspline/IR_LO_gfilter-1.),label ='Bsplin-Gfilter')
# plt.semilogx(km,100.*(IR_LO_bspline/IR_LO_dst-1.),label ='Bsplin-DST')
# plt.semilogx(km,100.*(IR_LO_gfilter/IR_LO_dst-1.),label ='Gfilter-DST')
# plt.xlim(1.e-3,1.)
# plt.minorticks_on()
# plt.text(0.005, -0.3, 'IR-res, LO', fontsize=13)
# plt.xlabel('$k \; [h/\mathrm{Mpc}]$')
# plt.ylabel('frac diff [\%]')
# plt.legend()
# plt.savefig("frac_diff_LO.pdf")
# plt.show()

# plt.semilogx(km,100.*(IR_NLO_bspline/IR_NLO_gfilter-1.),label ='Bsplin-Gfilter')
# plt.semilogx(km,100.*(IR_NLO_bspline/IR_NLO_dst-1.),label ='Bsplin-DST')
# plt.semilogx(km,100.*(IR_NLO_gfilter/IR_NLO_dst-1.),label ='Gfilter-DST')
# plt.xlim(1.e-3,1.)
# plt.minorticks_on()
# plt.text(0.005, -0.08, 'IR-res, NLO', fontsize=13)
# plt.xlabel('$k \; [h/\mathrm{Mpc}]$')
# plt.ylabel('frac diff [\%]')
# plt.legend()
# plt.savefig("frac_diff_NLO.pdf")
# plt.show()

```
